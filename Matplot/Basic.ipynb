```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5, 11)
y = x**2
```

## Figure with multiple axes

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.5, 0.5])
axes2 = fig.add_axes([0.7, 0.1, 0.5, 0.5])

axes1.plot(x,y)
axes1.set_title('X vs Y')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')

axes2.plot(y,x,'r')
axes2.set_title('Y vs X')
axes2.set_xlabel('Y')
axes2.set_ylabel('X')

```

## Create multiple axes with subplots. This is preffered method with higher abstraction

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('X vs Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

axes[1].plot(y,x,'r')
axes[1].set_title('Y vs X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

## Mention the figuresize of subplots

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (8,3))

axes[0].plot(x,y)
axes[0].set_title('X vs Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

axes[1].plot(y,x,'r')
axes[1].set_title('Y vs X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

## legend : 
* It needs a label
* Particularly seful when you have multiple plots in one axes

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.5, 0.5])

axes.plot(x,y, color ='r', label = 'X vs Y')
axes.plot(y,x, color ='b', label = 'Y vs X')
axes.legend(loc=0)
axes.set_title('Legend Example')
```

## Add marker and anotation

```python
x = np.linspace(0,10, 6)
y = x**2

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.5, 0.5])

axes.plot(x,y, color ='r', marker='o')
for xi, yi in zip(x, y):
    axes.annotate(f'{int(xi),int(yi)}', (xi, yi), textcoords="offset points", xytext=(0,10), ha='center')
```
