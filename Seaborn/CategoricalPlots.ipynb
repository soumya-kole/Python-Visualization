```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips
```

## Barplot

```python
sns.barplot(x='sex', y='total_bill', data=tips )
```

## Barplot with custom estimatot

```python
import numpy as np
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.sum )
```

## Box plot

```python
sns.boxplot(x='sex', y='total_bill', data=tips, hue='day')
```

```python
sns.boxplot(x='sex', y='total_bill', data=tips, hue='smoker')
```
