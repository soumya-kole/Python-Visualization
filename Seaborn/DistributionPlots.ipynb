```python
import seaborn as sns
```

## Seaborn has some builtin datasets. We can load one for our testing

```python
tips = sns.load_dataset('tips')
tips
```

## displot is distribution of one variable in a histogram

```python
sns.displot(tips['total_bill'])
```

## KDE for same distribution

```python
sns.kdeplot(tips['total_bill'])
```

## Histogram and KDE at the same time

```python
sns.distplot(tips['total_bill'])
```

## Jointplot is distribution of two variables one against another

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

## Jointplot with a regression line

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

## Pairplot creates a grid of scatterplots and histograms for all pairs of numerical values within a dataset
* You can use hue to pass categorical columns

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex')
```
